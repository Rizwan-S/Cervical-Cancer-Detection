# Mapping classes to numbers in a dictionary

```python
bethesda = {
    "Negative for intraepithelial lesion": 0,
    "ASC-US": 1,
    "LSIL": 2,
    "ASC-H": 3,
    "HSIL": 4,
    "SCC": 5
}
```

# Create Label Files

```python
counter = 0

for index, row in df.iterrows():
  filePath = ""
  if(row['image_filename'] in trainImgs):
    filePath = trainLabelsPath + os.path.splitext(row['image_filename'])[0] + '.txt'
  elif(row['image_filename'] in testImgs):
    filePath = testLabelsPath + os.path.splitext(row['image_filename'])[0] + '.txt'
  elif(row['image_filename'] in valImgs):
    filePath = valLabelsPath + os.path.splitext(row['image_filename'])[0] + '.txt'
  
  class_id = bethesda[row['bethesda_system']]
  center_x = row['nucleus_x'] /1376
  center_y = row['nucleus_y'] /1020
  bbox_width = 100 / 1376
  bbox_height = 100 / 1020
  towrite = f'{class_id} {center_x} {center_y} {bbox_width} {bbox_height}\n'

  fp = open(filePath, "a")
  fp.write(towrite)
  fp.close()

  counter = counter + 1
  if(counter % 100 == 0):
    print(counter)
```
