```python
trainImgsPath = "/content/drive/MyDrive/Cervix Cytology Project/custom_dataset_ttv/images/train/"
testImgsPath = "/content/drive/MyDrive/Cervix Cytology Project/custom_dataset_ttv/images/test/"
valImgsPath = "/content/drive/MyDrive/Cervix Cytology Project/custom_dataset_ttv/images/val/"

csvFile = "/content/drive/MyDrive/Cervix Cytology Project/custom_dataset_ttv/classifications.csv"
```

```python
df = pd.read_csv(csvFile)
df
```

```python
trainImgs = os.listdir(trainImgsPath)
testImgs = os.listdir(testImgsPath)
valImgs = os.listdir(valImgsPath)

len(trainImgs), len(testImgs), len(valImgs)
```

```python
trainCounts = {}
testCounts = {}
valCounts = {}

toCheck = ['SCC']

for index, row in df.iterrows():
  if(row['bethesda_system'] in toCheck):

    if(row['image_filename'] in trainImgs):
      if row['image_filename'] not in trainCounts:
         trainCounts[row['image_filename']] = 0
      trainCounts[row['image_filename']] = trainCounts[row['image_filename']] + 1

    elif(row['image_filename'] in testImgs):
      if row['image_filename'] not in testCounts:
        testCounts[row['image_filename']] = 0
      testCounts[row['image_filename']] = testCounts[row['image_filename']] + 1
    elif(row['image_filename'] in valCounts):
      if row['image_filename'] not in valCounts:
        valCounts[row['image_filename']] = 0
      valCounts[row['image_filename']] = valCounts[row['image_filename']] + 1
```
